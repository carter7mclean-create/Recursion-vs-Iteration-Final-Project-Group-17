# Recursion vs Iteration Poster Benchmarks

This notebook supports the final research poster for the project.

**Research goal:** determine when recursive implementations become less practical than iterative ones in Java because of call-stack overhead, runtime overhead, or `StackOverflowError`.

This notebook can either:

- load the benchmark results already generated in the repo, or
- rerun the Java benchmarks and regenerate the CSV and charts.


```python
from pathlib import Path
import sys

def find_repo_root(start: Path) -> Path:
    for candidate in [start, *start.parents]:
        if (candidate / 'src').is_dir() and (candidate / 'README.md').exists():
            return candidate
    raise RuntimeError('Could not locate the repository root from the current working directory.')

ROOT = find_repo_root(Path.cwd())
ANALYSIS_DIR = ROOT / 'results' / 'analysis'
if str(ANALYSIS_DIR) not in sys.path:
    sys.path.insert(0, str(ANALYSIS_DIR))

import poster_benchmarks as pb

ROOT
```

```python
RUN_BENCHMARKS = False
NUM_RUNS = 10
TIMEOUT_SECONDS = 1200

if RUN_BENCHMARKS or not pb.CSV_PATH.exists():
    result = pb.run_pipeline(num_runs=NUM_RUNS, timeout_seconds=TIMEOUT_SECONDS)
    rows = result['rows']
    chart_paths = result['chart_paths']
else:
    rows = pb.load_results()
    chart_paths = pb.generate_charts(rows)

print(f'CSV: {pb.CSV_PATH}')
for chart_path in chart_paths:
    print(f'Chart: {chart_path}')
```

```python
from IPython.display import HTML, Markdown, SVG, display

display(Markdown('## Benchmark Summary'))
display(HTML(pb.summary_html(rows)))
```

```python
display(Markdown('## Poster Charts'))
for chart_path in chart_paths:
    display(Markdown(f'### {chart_path.name}'))
    display(SVG(filename=str(chart_path)))
```
